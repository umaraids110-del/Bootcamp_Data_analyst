```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\SYNCNER TRAINING\BOOTCAMP\Day 3\Task 2\Income.csv")
df
```

```python
print(df.head())
print(df.columns)
```

```python
df["Average"]=df.mean(axis=1)
```

```python
plt.figure(figsize=(5,5))
```

```python
plt.plot(df["Average"],df.index,marker="o",linestyle="--",color="Green",linewidth=1,label="Average")
plt.ylabel("Year",fontsize=12)
plt.xlabel("Average Income(Years)",fontsize=12)
plt.title("Income trends over time",fontsize=14,fontweight="bold")
plt.grid(True,linestyle="--",alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
