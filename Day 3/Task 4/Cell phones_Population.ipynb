```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
population_data = pd.read_csv(r"D:\SYNCNER TRAINING\BOOTCAMP\Day 3\Task 3\population.csv")
cell_phones_data = pd.read_csv(r"D:\SYNCNER TRAINING\BOOTCAMP\Day 3\Task 3\cell_phones_per_100_people.csv.csv")
```

```python
df_population_numeric = population_data.set_index("country").apply(lambda col: col.map(lambda x: pd.to_numeric(str(x).replace("M", "e6").replace("k", "e3"), errors='coerce')))
global_population = df_population_numeric.sum(axis=0)
df_population_numeric 
```

```python
df_cell_phones_numeric = cell_phones_data.set_index("country").apply(lambda col: col.map(lambda x: pd.to_numeric(str(x).replace("M", "e6").replace("k", "e3"), errors='coerce')))
global_cell_phones = df_cell_phones_numeric.sum(axis=0)
df_cell_phones_numeric
```

```python
cell_phones_per_person = (global_cell_phones / global_population).dropna()
cell_phones_per_person_df = cell_phones_per_person.reset_index()
cell_phones_per_person_df.columns = ["Year", "Cell Phones per Person"]
plt.figure(figsize=(6, 4))
plt.plot(cell_phones_per_person_df["Cell Phones per Person"], cell_phones_per_person_df["Year"], marker="o", linestyle="-", color="b")
plt.xlabel("Number of Cell Phones per Person")
plt.ylabel("Year")
plt.title("Rise of Cell Phones per Person Over Time")
plt.grid(True)
plt.show()
```
