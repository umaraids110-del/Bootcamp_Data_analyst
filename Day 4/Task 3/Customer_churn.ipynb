```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\SYNCNER TRAINING\BOOTCAMP\Day 4\Task 3\Teleco_churn_data.csv')
df
```

```python
df['Total Charges'] = pd.to_numeric(df['Total Charges'], errors='coerce')
df['Total Charges'] = df['Total Charges'].fillna(df['Total Charges'].median())
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
corr_matrix = df[numeric_cols].corr()
churn_corr = corr_matrix['Churn Value'].drop('Churn Value').sort_values(ascending=False)
print("Top features positively correlated with churn:")
print(churn_corr[churn_corr > 0].head(5))
print("\nTop features negatively correlated with churn:")
print(churn_corr[churn_corr < 0].head(5))
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
