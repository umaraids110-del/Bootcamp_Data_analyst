```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Solution.csv')
df
```

```python
df.columns
```

```python
df.info()
```

```python
df["Average"]=df.mean(axis=1)
plt.figure(figsize=(5,5))
```

```python
plt.plot(df["Average"],df.index,marker="o",linestyle="--",color="blue",linewidth=1,label="Average")
plt.ylabel("Year",fontsize=12)
plt.xlabel("Average life expectancy(Years)",fontsize=12)
plt.title("Life expectancy trends over time",fontsize=14,fontweight="bold")
plt.grid(True,linestyle="--",alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
