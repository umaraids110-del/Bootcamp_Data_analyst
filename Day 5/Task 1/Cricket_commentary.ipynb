```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
# Load dataset (use 'r' to handle file path correctly)
df = pd.read_csv(r'D:\SYNCNER TRAINING\BOOTCAMP\Day 5\Task 1\Cricket_highlights.csv')
df
```

```python
print(df.head())
```

```python
# Extract the 'Commentary' column and combine all text into a single string
commentary_text = " ".join(df["Commentary"].dropna().astype(str))
```

```python
# Define stopwords to remove common words
stopwords = set(STOPWORDS)
```

```python
# Generate Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords).generate(commentary_text)
```

```python
# Display Word Cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # HideÂ axes
plt.show()
```
